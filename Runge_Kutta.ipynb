```python
import numpy as np
from math import e
import matplotlib.pyplot as plt

t = 0
tf = 0.3
h = 0.1
n = round((tf - t)/h)
u = 2

def aproximada(u,t):
  valores = []
  for i in range(n):
    f1 = - u + t + 2
    uf2 = u + (h/2)*f1
    tf2 = t + (h/2)
    f2 = - uf2 + tf2 + 2
    uf3 = u + (h/2)*f2
    tf3 = t + (h/2)
    f3 = - uf3 + tf3 + 2
    uf4 = u + (h)*f3
    tf4 = t + h
    f4 = - uf2 + tf2 + 2
    u = u + h*(f1 + 2*f2 + 2*f3 + f4)/6
    t = t + h
    valores.append(u)
  return valores

print(aproximada(u,t))

def exata(t):
  return t + 1 +(1/e**t)

print(exata(tf))

# Lista da solução aproximada
valores_aproximados = aproximada(u, t)

# Lista da solução exata
valores_exatos = []
for i in range(n):
  valores_exatos.append(exata(tf))

# Plotagem dos gráficos
iteracoes = np.linspace(t, tf, n, endpoint=False)
plt.plot(iteracoes, valores_aproximados, 'r-', label='Solução Aproximada')
plt.plot(iteracoes, valores_exatos, 'b-', label='Solução Exata')
plt.legend()
plt.xlabel('Iteraçoes')
plt.ylabel('Valor de u')
plt.title('Comparação entre Solução Aproximada e Exata')
plt.show()
```
